```python
x = np.linspace(-3,3,100)
y=[0,0,0,0]
y[0]=np.sin(x)
y[1]=np.cos(x)
y[2]=1/(1+np.exp(-x))
y[3]=np.exp(x)

figure, ax = plt.subplots(2, 2)
i=0
for a in range(len(ax)):        # 0 ,1
    for b in range(len(ax[a])): # a = 0 일때 b =0, 1 반복 / a = 1 일 때 b = 0, 1 반복
        ax[a, b].plot(x, y[i])
        subplot_title=("Subplot"+str(i))
        ax[a,b].set_title(subplot_title)
        i=i+1

figure.tight_layout()
plt.show()
```
